```python
var = set(list(map(lambda x: x[:-1], df.columns)))
```

```python
rounding = {
    'mdage':1,
    'avghhsz':2,
    'avgfmsz':2,
    'mntrvtm':1,
    'hovacrt':1,
    'mdrms':1,
    'avghhsooc':2,
    'avghhsroc':2
}
r = []
for i in ['mdage']: 
    cols = ['geo_id', 'name', i+'c', i+'e', i+'m', i+'p', i+'z']
    dff = df.loc[:, cols]
    dff.columns=['geoid', 'geoname', 'c', 'e', 'm', 'p', 'z']
    dff['variable'] = i
    dff.loc[:,'c'] = dff['c'].round(1)
    dff.loc[:,'e'] = dff['e'].round(rounding.get(i, 0))
    dff.loc[:,'m'] = dff['m'].round(rounding.get(i, 0))
    dff.loc[:,'p'] = dff['p'].round(1)
    dff.loc[:,'z'] = dff['z'].round(1)
    r.append(dff)
```

```python
result = pd.concat(r)
```
